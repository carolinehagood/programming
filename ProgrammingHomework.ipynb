```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 50, endpoint=False)[1:]
log_values = np.log(x)
exp_values = np.exp(x)
```

```python
plt.scatter(x, log_values, color='orange')
plt.scatter(x, exp_values, color='blue')
```

```python
plt.scatter(x, log_values, color='orange')
plt.scatter(x, exp_values, color='blue')

plt.xlabel('x')
plt.ylabel('Function Values')
plt.title('Exponential and Natural Log Functions')
```

```python
plt.scatter(x, log_values, label='Natural Log', color='orange')
plt.scatter(x, exp_values, label='Exponential', color='blue')

plt.xlabel('x')
plt.ylabel('Function Values')
plt.title('Exponential and Natural Log Functions')

plt.legend(loc='lower right')
plt.show()
```

Create a grid of 100 equally spaced points ranging from -6.5 to 6.5.
Use the sine and cosine functions in Numpy to compute the values of those functions for each point on your grid. (You'll have to find out what those functions are.)
Plot the values of the two functions for the values on the grid on the same plot.
The scatter plot looks really noisy. Instead of plt.scatter(x,y) to make a scatter plot, use plt.plot(x,y) to make a line graph.
Make the plot again, with labels for the axes, a title, and a legend in the lower left instead of the lower right.

```python
x1 = np.linspace(-6.5, 6.5, 100, endpoint=False)
sin_values = np.sin(x1)
cos_values = np.cos(x1)
```

```python
plt.scatter(x1, sin_values, color='red')
plt.scatter(x1, cos_values, color='green')
```

```python
plt.plot(x1, sin_values, color='red')
plt.plot(x1, cos_values, color='green')
```

```python
plt.plot(x1, sin_values, label='Sine', color='red')
plt.plot(x1, cos_values, label='Cosine', color='green')

plt.xlabel('X')
plt.ylabel('Function Values')
plt.title('Cosine and Sine Functions')
plt.legend(loc='lower left')

plt.show()
```
